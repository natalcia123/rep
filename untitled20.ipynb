```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

# Funkcje do interpolacji
def f1(x):
    return np.sin(x)

def f2(x):
    return np.sin(1/x)

def f3(x):
    return np.sign(np.sin(8*x))

# Funkcje jÄ…drowe
def h1(t):
    return np.where((t >= 0) & (t < 1), 1, 0)

def h2(t):
    return np.where((t >= -1/2) & (t <= 1/2), 1, 0)

def h3(t):
    return np.where((t >= -1) & (t <= 1), 1 - np.abs(t), 0)

 # WybÃ³r funkcji do interpolacji
functions = [f1, f2, f3]

# WybÃ³r jÄ…der do konwolucji
kernels = [h1, h2, h3]

# WybÃ³r liczby punktÃ³w
n_points_list = [10, 20, 100]



# Interpolacja funkcji za pomocÄ… jedno-wymiarowej konwolucji
def interpolate_function(func, kernel, n_points):
    x_original = np.linspace(-np.pi, np.pi, n_points)
    y_original = func(x_original)

    x_interpolated = np.linspace(-np.pi, np.pi, n_points)
    y_interpolated = convolve(y_original, kernel, mode='same', method='direct' ) / np.sum(kernel)
    y_interpolated = np.interp(x_interpolated, x_original,y_interpolated)

    mse = np.mean((func(x_interpolated) - y_interpolated)**2)  # Obliczanie MSE
    return x_interpolated, y_interpolated, mse


# Interpolacja i obliczenie MSE
for func in functions:
    print(f"Interpolacja funkcji {func.__name__}:")

    for kernel in kernels:
        print(f"JÄ…dro {kernel.__name__}:")

        plt.plot(np.linspace(-np.pi, np.pi, 1000), func(np.linspace(-np.pi, np.pi, 1000)), label="OryginaÅ‚", linestyle='dashed')

        for n_points in n_points_list:
            x_interpolated, y_interpolated, mse = interpolate_function(func, kernel(np.linspace(-1, 1, n_points)), n_points)
            plt.plot(x_interpolated, y_interpolated, label=f"Interpolacja ({n_points} pkt), MSE: {mse:.4f}")

            # Dodanie oznaczenia punktÃ³w interpolacji
            #plt.scatter(x_interpolated, y_interpolated, marker='o')

        plt.legend()
        plt.show()
```
